write a program to print fizz if numbers in the range from 1 to 50 (both inclusive) is div by 3 , buzz if numbers are div by 5 and fizz buzz if num is div by both.

```python
for i in range(1,51):
    if i%3==0 and i%5==0:
        print('fizzbuzz ',i)
    elif i%3==0:
        print('fizz ',i)
    elif i%5==0:
        print('buzz ',i)
    else:
        print(i)
```
